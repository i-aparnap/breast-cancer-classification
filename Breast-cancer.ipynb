```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Load dataset
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
X
```

```python
# Check for missing values
X.isnull().sum()
```

```python
# Apply Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

Justification: The breast cancer dataset from sklearn contains no missing values, so no imputation is needed.
Always necessary to check for nulls before applying any machine learning model, as most models can't handle missing data.
The dataset contains features like mean radius, mean texture, worst area, etc., which are on different scales.
StandardScaler brings all features to a similar scale (mean = 0, std = 1), which is especially important for algorithms like:
Logistic Regression,KNN etc.

```python
#Q2.
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Split the scaled data
y = pd.Series(data.target, name='target')
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
```

Logistic regression models the probability that a given input belongs to a certain class using the sigmoid function.
Why suitable:Fast and interpretable.Performs well on linearly separable data like this dataset



```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
```

Builds a tree-like model by splitting data into subsets based on feature values that maximize information gain
Why suitable:Captures non-linear relationships,Easy to interpret

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

An ensemble of decision trees using bagging (random sampling) and majority voting for prediction.
Why suitable:High accuracy
Reduces overfitting compared to a single tree
Handles both linear and non-linear patterns

```python
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
```

Finds the optimal hyperplane that maximizes the margin between classes.
Why suitable:Effective in high-dimensional spaces
Works well with small to medium-sized datasets
Sensitive to feature scaling (hence scaling was important)



```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("k-NN Accuracy:", accuracy_score(y_test, y_pred_knn))
```

Classifies a point based on the majority class of its k nearest neighbors in the feature space.
Why suitable:
Simple and intuitive
Works well with well-separated classes
Performance depends on proper scaling and choosing k
